```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import urllib.request
import json
import re
import numpy as np
```

```python
class Movie :
    def __init__(self,json_object):
        self.title = json_object['title']
        self.imdb_code = json_object['imdb_code']
        self.year = json_object['year']
        self.imdbRat = json_object['rating']
        self.mpaa = json_object['mpa_rating']
        self.genres = json_object['genres']
        self.runtime = json_object['runtime']
        self.language = json_object['language']
        self.budget = -1
        self.gross = -1
        self.countries = []
        self.prod = ["Unknown"]
        self.starpower = 0
        self.director= ""
        self.director_starpower = 0
```

```python
browser = webdriver.Chrome("C:/Users/rafaa/Desktop/Projects/Python/chromedriver.exe")
browser.set_page_load_timeout(20)
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.3'}
```

```python
star_LUT = {}
scrapped_movies = np.array([],dtype=Movie)
```

```python
movies_list = []
for page in range(1,200):
    yts_url = "https://yts.am/api/v2/list_movies.json?limit=50&page="+str(page)
    req = urllib.request.Request(url=yts_url, headers=headers)
    json_file = urllib.request.urlopen(req).read()
    parsed_json = json.loads(json_file)
    for movie in parsed_json['data']['movies'] :
        movies_list.append(Movie(movie))

np.save("movielist" + str(time.time()),np.array(movies_list))
#movies_list = np.load("movielist1554673158.2307246.npy")
for movie in movies_list:
    print(movie.title)
```

```python
def getStarPower(star_pro_url):
    if(star_pro_url in star_LUT):
        print("Star already exists, using LUT")
        return star_LUT[star_pro_url]
    nb_power = 0;
    print(star_pro_url);
    result =0
    browser.get(star_pro_url)
    browser.execute_script("document.getElementById('logged_out_upsell').style.display = 'none';")
    #Past Film & Video
    try:
        past_filmography = browser.find_elements_by_xpath("//*[contains(text(), 'Past Film & Video')]/ancestor::div")
        filmography_box = past_filmography[-3]
    except:
        past_filmography = browser.find_elements_by_xpath("//*[contains(text(), 'Self')]/ancestor::div")
        filmography_box = past_filmography[-3]

    expanders = filmography_box.find_elements_by_class_name("a-expander-prompt")
    
    if(len(expanders)>0):
        print(len(expanders))
        try:
            for expander in expanders:
                expander.click()
        except:
            print("cannot click")
    try :
        past_movies = filmography_box.find_element_by_id("past_film_filmography");
    except:
        past_movies = filmography_box.find_element_by_id("self_filmography");
    rows = past_movies.find_elements_by_tag_name("tr");
    
    for row in rows :
        if(row.get_attribute('data-filter-item-id')):
            value = row.find_elements_by_tag_name('td')[4].text
            in_dollars = convertTextToDollars(value)
            result += in_dollars
            if(in_dollars>0):
                nb_power+=1
    if(nb_power>0):
        result = result/nb_power
    star_LUT[star_pro_url] =result
    return result

def convertTextToDollars(string):
    if(string.find('$')==-1):
        return 0
    numerals = string.split("$")[1]
    unit = " ".join(re.findall("[a-zA-Z]+", numerals))
    digits= float(" ".join(re.findall("[0-9\.]+", numerals)))
    
    if(unit is "B"):
        return digits * 10000000
    elif(unit == "MM"):
        return digits*1000000
    elif(unit is "K"):
        return digits * 1000
    else:
        return digits

#getStarPower("https://pro.imdb.com/name/nm0001060")
getStarPower("https://pro.imdb.com/name/nm5065920/?ref_=tt_cl_i3")

```

```python
filtered_movies = list(filter(lambda mv : (int(mv.year)>1999) and mv.language == "English",movies_list))
i=2000
last_ignored = ""
for movie in filtered_movies :
    print(str(i)+" --------------")
    i = i+1
    tries_left = 2
    for try_index in range(tries_left) : 
        if(last_ignored == movie.title):
            continue
        last_ignored = ""
        print("Try number " + str(try_index+1))
        try:
            imdb_url = "https://www.imdb.com/title/" + movie.imdb_code
            browser_tries_left = 4
            for browser_try_index in range(browser_tries_left):
                try:
                    browser.get(imdb_url)
                    break
                except:
                    print("Browser timeout")
            print("------Scrapping : " +movie.imdb_code + "   "+movie.title)
            #USA Gross
            country_elt = browser.find_elements_by_xpath("//*[contains(text(), 'Country')]/ancestor::div")
            if(len(country_elt)>0):
                country = list(map(lambda s:s.strip(),country_elt[-1].text.split(':')[1].split('|')))
            else:
                country = []
                last_ignored = movie.title
                print("No country was found")
                continue
            movie.countries = country
            if(not("USA" in movie.countries or "UK" in movie.countries)):
                last_ignored = movie.title
                print("NOT USA/UK, ignoring")
                continue

            #Budget
            budget_elt = browser.find_elements_by_xpath("//*[contains(text(), 'Budget')]/ancestor::div")
            if(len(budget_elt)>0):
                budget = int(budget_elt[-1].text.split('$')[1].split('(estimated)')[0].replace(',',''))
            else:
                budget =  -1
            movie.budget = budget

            #USA Gross
            gross_elt = browser.find_elements_by_xpath("//*[contains(text(), 'Gross USA')]/ancestor::div")
            if(len(gross_elt)>0):
                gross = int(gross_elt[-1].text.split('$')[1].split(' ')[0].replace(',',''))
            else:
                gross = -1
            movie.gross = gross


            #Runtime
            if(movie.runtime == 0):
                runtime_elt = browser.find_elements_by_xpath("//*[contains(text(), 'Runtime')]/ancestor::div")
                if(len(runtime_elt)>0):
                    runtime = runtime_elt[-1].text.split(':')[1].split(' ')[0]
                else:
                    runtime = -1
                movie.runtime = runtime


            #Production Company
            prod_elt = browser.find_elements_by_xpath("//*[contains(text(), 'Production Co')]/ancestor::div")
            if(len(prod_elt)>0):
                prod = list(map(lambda x:x.strip(),prod_elt[-1].text.split(':')[1].split('See more')[0].split(',')))
            else:
                prod = ["Unknown"]
            movie.prod = prod

            #Director
            director_elts = browser.find_element_by_id("main_top").find_elements_by_class_name("credit_summary_item")
            cast_list = browser.find_elements_by_css_selector(".cast_list .primary_photo a")
            cast_list_urls = list(map(lambda x: x.get_attribute("href").replace("www","pro"),cast_list[:3]))
            director_elt = list(filter(lambda x:x.text.find("Director:")>-1,director_elts))
            if(len(director_elt)>0):
                director = director_elt[0].text.split(":")[1].strip().split("(")[0]
                director_starpower = getStarPower(director_elt[0].find_elements_by_tag_name("a")[0].get_attribute("href").replace("www","pro"))
            else:
                director = "Unknown"
                director_starpower = 0
            movie.director = director
            movie.director_starpower = director_starpower



            #Cast List
            movie.starpower = 0 #TODO remove 
            
            for cast_elt in cast_list_urls:
                movie.starpower += getStarPower(cast_elt)/3
            print(movie.title + "  Star power :" + str(movie.starpower))
            scrapped_movies = np.append(scrapped_movies,movie)
            print("Completed "+ movie.title)
            break

        except Exception as e:
            print("An error occured while scrapping '" + movie.title+"' , Skipping this movie");
            print(e)
            
```
